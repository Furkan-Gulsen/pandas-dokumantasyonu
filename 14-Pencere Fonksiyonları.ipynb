# PYTHON Pandas - Pencere Fonksiyonları (Window Functions)
Sayısal veriler üzerinde çalışmak için pandalar, pencere istatistikleri için haddeleme, genişletme ve katlanarak hareket eden ağırlıklar gibi birkaç varyant sağlar. Bunlar arasında toplam, ortalama, medyan, varyans, kovaryans, korelasyon vb.vardır.
Şimdi bunların her birinin DataFrame nesnelerine nasıl uygulanabileceğini öğreneceğiz.

## .rolling() Fonksiyonu
Bu işlev bir dizi veri üzerinde uygulanabilir. Window=n bağımsız değişkenini belirtir ve bunun üzerine uygun istatistiksel işlevi uygular.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10, 4),
                  index = pd.date_range('1-1-2000', periods=10),
                  columns = ['A', 'B', 'C', 'D'])
df.rolling(window=3).mean()
```

**Not**
Pencere boyutu 3 olduğundan, ilk iki öğe için boş değerler vardır ve üçüncü değer n, n-1 ve n-2 elemanlarının ortalaması olacaktır. Böylece yukarıda belirtildiği gibi çeşitli işlevleri de uygulayabiliriz.

## expanding() Fonksiyonu
Bu işlev bir dizi veri üzerinde uygulanabilir. Min_periods=n bağımsız değişkenini belirtir ve bunun üzerine uygun istatistiksel işlevi uygular.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10, 4),
   index = pd.date_range('1/1/2000', periods=10),
   columns = ['A', 'B', 'C', 'D'])
df.expanding(min_periods=3).mean()
```

## .ewm() Fonksiyonu
ewm bir dizi veri üzerine uygulanır. Com, span, halflife bağımsız değişkeninden herhangi birini belirtir ve bunun üzerine uygun istatistiksel işlevi uygular. Ağırlıkları katlanarak atar.

```python
import pandas as pd
import numpy as np
 
df = pd.DataFrame(np.random.randn(10, 4),
   index = pd.date_range('1/1/2000', periods=10),
   columns = ['A', 'B', 'C', 'D'])
df.ewm(com=0.5).mean()
```

Pencere işlevleri, eğriyi düzleştirerek veri içindeki eğilimleri grafiksel olarak bulmada büyük ölçüde kullanılır. Günlük verilerde çok fazla varyasyon varsa ve çok sayıda veri noktası mevcutsa, örnekleri almak ve çizmek bir yöntemdir ve pencere hesaplamalarını uygulamak ve sonuçları grafiğe çizmek başka bir yöntemdir. Bu yöntemlerle eğriyi veya eğilimi düzeltebiliriz.
