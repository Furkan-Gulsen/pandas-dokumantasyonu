# Temel Çizim: Plot
Seriler ve DataFrame üzerindeki bu işlevsellik, matplotlib kütüphaneleri plot() yönteminin etrafında sadece basit bir sarıcıdır.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10,4),
                  index=pd.date_range('1/1/2000',periods=10),
                  columns=list('ABCD'))

df.plot()
```

X ve y anahtar kelimelerini kullanarak bir sütunu diğerine çizebiliriz.
Çizim yöntemleri, varsayılan çizgi çiziminden başka bir avuç çizim stiline izin verir. Bu yöntemler, plot () için tür anahtar kelime bağımsız değişkeni olarak sağlanabilir.
- bar parselleri için bar veya barh
- histogram için hist
- boxplot için box
- alanlar için plots

## Bar Plot
Şimdi bir bar grafiği oluşturarak ne olduğunu görelim. Bir çubuk grafik aşağıdaki şekilde oluşturulabilir.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df.plot.bar()
```

Tek bir çubukta oluşturmak için:

```python
import pandas as pd
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df.plot.bar(stacked=True)
```

Yatay çubuk grafikleri elde etmek için barh yöntemini kullanın.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])

df.plot.barh(stacked=True)
```

## Histogram
Histogramlar grafik kullanılarak çizilebilir.hist () yöntemine kutuları sayısını belirtebilirsiniz.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a':np.random.randn(1000)+1,
                   'b':np.random.randn(1000),
                   'c':np.random.randn(1000) - 1},
                   columns=['a', 'b', 'c'])

df.plot.hist(bins=20)
```

Her sütun için farklı histogramları çizmek için aşağıdaki kodu kullanın.

```python
import pandas as pd
import numpy as np

df=pd.DataFrame({'a':np.random.randn(1000)+1,
                 'b':np.random.randn(1000),
                 'c':np.random.randn(1000) - 1},
                 columns=['a', 'b', 'c'])

df.hist(bins=20)
```

## Box Plots
Boxplot Series.box.plot () ve DataFrame.box.plot () veya DataFrame.boxplot () her sütun içindeki değerlerin dağılımını görselleştirmek içindir.


```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df.plot.box()
```

## Area Plot

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot.area()
```

## Scatter Plot

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
df.plot.scatter(x='a', y='b')
```

## Pie Chart

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(3 * np.random.rand(4),
                  index=['a', 'b', 'c', 'd'],
                  columns=['x'])
df.plot.pie(subplots=True)
```
