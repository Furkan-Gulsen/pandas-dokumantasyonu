# PYTHON Pandas - Sıralama
Pandaslarda iki tür sıralama vardır.Bunlar:
- etikete göre sıralama
- değere göre sıralama

Bir çıktı ile bir örnek düşünelim.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10,2),
                        index=[1,4,6,2,3,5,9,8,0,7],
                        columns = ['col2','col1'])
df
```

## Etikete Göre Sıralama
Sort_index() yöntemini kullanarak, eksen bağımsız değişkenlerini ve sıralama sırasını geçirerek, DataFrame sıralanabilir. Varsayılan olarak, sıralama, satır etiketlerinde artan düzende yapılır.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10,2),
                  index=[1,4,6,2,3,5,9,8,0,7],
                  columns = ['col2','col1'])

siralama = df.sort_index()
siralama
```

### Sıralama sırası
Boolean değerini artan parametreye geçirerek, sıralama sırası kontrol edilebilir. Bunu anlamak için aşağıdaki örneği ele alalım.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10,2),
                  index=[1,4,6,2,3,5,9,8,0,7],
                  columns = ['col2','col1'])

siralama = df.sort_index(ascending=False)
siralama
```

## Değere Göre Sıralama
Yukarıdaki yapıya benzer bir kullanımı vardır. Örneği inceleyerek yapıyı daha iyi anlayalım.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'col1':[2,1,1,1],'col2':[1,3,2,4]})
siralama = df.sort_values(by='col1')

siralama
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'col1':[2,1,1,1],'col2':[1,3,2,4]})
siralama = df.sort_values(by='col2')

siralama
```

İki sütunu kullanarakta bu sıralamayı sağlayabiliriz.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'col1':[2,1,1,1],'col2':[1,3,2,4]})
siralama = df.sort_values(by=['col1','col2'])

siralama
```
